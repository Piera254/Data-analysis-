```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
data = pd.read_csv("Documents/customer_data.csv")
print(data.head())
```

```python
print(type(data))
```

```python
#printing the first five columns of the data
print(data.head())
```

```python
data.describe()
```

```python
#using this function to check for missing values
data.info()
```

```python
#distribution of the age of customers
plt.figure(figsize =(10,8))
sns.countplot( x="age", hue ="gender", data = data)
```

```python
#visualizing loyalty statuus
sns.countplot(x = "loyalty_status", data = data)
```

```python
#a more visual plot of the loyalty_status of customers
sns.countplot(x = "loyalty_status", hue = "education", data = data)
```

```python
#gender countplot
sns.countplot( x ="gender", data = data)
```

```python
#education level count plot
sns.countplot( x ="education", data = data)
```

```python
#loyalty status versus gender
sns.countplot( x ="loyalty_status", data = data, hue ="gender")
```

```python
#visualizing the purchase frequency
sns.countplot( x ="product_category", data = data, hue = "purchase_frequency")
```

```python
sns.countplot(x= "region", data = data, hue ="loyalty_status")
```

```python
#visualizing income distribution for both genders
sns.catplot( x='gender', y='income', data=data, kind='boxen')
```

```python
#selecting numeric data to calculate correlation
numeric_data = data.select_dtypes(include = "number")
```

```python
numeric_data.corr()
```

```python
#visualizing correlation
sns.heatmap(numeric_data.corr(), cmap = "coolwarm")
```

```python
#a hist plot of the age of custiomers
sns.histplot(data["age"], kde = True, bins = 20)
```

```python
#generating a pairplot
sns.pairplot(data)
```

```python
#a regression plot of income and the purchase amount
plt.figure(figsize=(10,6))
sns.lmplot(x = "income", y= "purchase_amount", hue = "gender", data = data)
```

```python
#customizing the regression plot for better insights
sns.lmplot( x= "income", y = "purchase_amount", hue = "gender", data = data, markers =['o', 'v'], scatter_kws = {'s':10})
```

```python
#further customization of the regression plot
sns.lmplot(x='purchase_amount',y='income',data=data,col='gender', row = "loyalty_status", scatter_kws ={"s":10})
```
